```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
df = pd.read_csv("../data/processed/data_understanding.csv",parse_dates=["InvoiceDate"])
```

```python
#Basic Information
```

```python
print("=" * 60)
print("DATASET BEFORE CLEANING")
print("=" * 60)

print(f"Rows    : {df.shape[0]:,}")
print(f"Columns : {df.shape[1]}")

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
#Remove Duplicate Rows
```

```python
duplicate_count = df.duplicated().sum()

print(f"Duplicate rows: {duplicate_count:,}")
```

```python
df = df.drop_duplicates()
```

```python
print(df.shape)
```

```python
#Handle Missing Customer IDs
```

```python
df["Customer ID"].isnull().sum()
```

```python
df = df.dropna(subset=["Customer ID"])
```

```python
df["Customer ID"].isnull().sum()
```

```python
#Remove Missing Descriptions
```

```python
df["Description"].isnull().sum()
```

```python
df = df.dropna(subset=["Description"])
```

```python
df["Customer ID"].dtype
```

```python
#Convert Customer ID to Integer
```

```python
df["Customer ID"] = df["Customer ID"].astype(int)
```

```python
#Remove Cancelled Orders
```

```python
cancelled = df["Invoice"].astype(str).str.startswith("C").sum()

print(cancelled)
```

```python
df = df[~df["Invoice"].astype(str).str.startswith("C")]
```

```python
#Remove Negative Quantities | invalid
```

```python
(df["Quantity"] <= 0).sum()
```

```python
df = df[df["Quantity"] > 0]
```

```python
#Remove Zero or Negative Prices | invalid
```

```python
(df["Price"] <= 0).sum()
```

```python
df = df[df["Price"] > 0]
```

```python
#Create Revenue Feature
```

```python
df["TotalAmount"] = df["Quantity"] * df["Price"]
```

```python
df.head()
```

```python
#Final Quality Check
```

```python
print("=" * 60)
print("DATASET AFTER CLEANING")
print("=" * 60)

print(f"Rows    : {df.shape[0]:,}")
print(f"Columns : {df.shape[1]}")

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
df.to_csv("../data/processed/cleaned_online_retail.csv",index=False)

print("✅ Clean dataset saved successfully!")
```
