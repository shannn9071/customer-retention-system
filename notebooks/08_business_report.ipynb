```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/customer_features.csv")

df.head()
```

```python
df.info()
```

```python
df.columns.tolist()
```

```python
# Customer counts
total_customers = len(df)

churned_customers = (df["Churn"] == 1).sum()

active_customers = (df["Churn"] == 0).sum()

churn_rate = (churned_customers / total_customers) * 100

# Customer metrics
average_frequency = df["Frequency"].mean()

average_monetary = df["Monetary"].mean()

average_order_value = df["AverageOrderValue"].mean()

average_customer_lifetime = df["CustomerLifetime"].mean()

average_purchase_gap = df["AveragePurchaseGap"].mean()

# Spending
highest_spending = df["Monetary"].max()

lowest_spending = df["Monetary"].min()

total_revenue = df["Monetary"].sum()
```

```python
top_customers = (
    df.sort_values("Monetary", ascending=False)
      .head(10)
)

top_customers
```

```python
print("="*60)

print("CUSTOMER RETENTION BUSINESS REPORT")

print("="*60)

print(f"Total Customers          : {total_customers}")

print(f"Active Customers         : {active_customers}")

print(f"Churned Customers        : {churned_customers}")

print(f"Customer Churn Rate      : {churn_rate:.2f}%")

print()

print(f"Total Revenue            : {total_revenue:,.2f}")

print(f"Average Monetary Value   : {average_monetary:.2f}")

print(f"Average Frequency        : {average_frequency:.2f}")

print(f"Average Order Value      : {average_order_value:.2f}")

print(f"Average Customer Life    : {average_customer_lifetime:.2f} days")

print(f"Average Purchase Gap     : {average_purchase_gap:.2f} days")

print(f"Highest Spending Customer: {highest_spending:.2f}")

print("="*60)
```

```python
top_customers.to_csv(
    "../reports/top_10_customers.csv",
    index=False
)
```

```python
report = f"""
# Customer Retention Business Report

## Executive Summary

Total Customers: {total_customers}

Active Customers: {active_customers}

Churned Customers: {churned_customers}

Customer Churn Rate: {churn_rate:.2f}%

---

## Revenue Metrics

Total Revenue: {total_revenue:,.2f}

Average Monetary Value: {average_monetary:.2f}

Highest Spending Customer: {highest_spending:.2f}

---

## Customer Behaviour

Average Purchase Frequency: {average_frequency:.2f}

Average Order Value: {average_order_value:.2f}

Average Customer Lifetime: {average_customer_lifetime:.2f} days

Average Purchase Gap: {average_purchase_gap:.2f} days

---

## Business Recommendations

1. Target churned customers with personalized offers.
2. Reward high-value customers through loyalty programs.
3. Re-engage customers with high recency using email campaigns.
4. Focus retention efforts on customers showing declining purchase frequency.
5. Monitor customer lifetime value to prioritize marketing investments.
"""

with open("../reports/business_report.md", "w", encoding="utf-8") as f:
    f.write(report)

print("Business report saved successfully!")
```
